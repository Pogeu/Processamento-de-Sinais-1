```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import chirp
from IPython.display import Audio
```

## Definir parâmetros

```python
fs = 44100
duracao = 5

t = np.linspace(0, duracao, int(fs*duracao), endpoint=False)

f0 = 500
f1 = 10000
```

## Gerar sinais chirp

```python
chirp_linear = chirp(t, f0=f0, f1=f1, t1=duracao, method='linear')

chirp_quadratic = chirp(t, f0=f0, f1=f1, t1=duracao, method='quadratic')

chirp_log = chirp(t, f0=f0, f1=f1, t1=duracao, method='logarithmic')
```

## Plotar sinais no domínio do tempo

```python
plt.figure(figsize=(12,6))

n = int(0.02*fs)  # mostrar primeiros 20 ms

plt.plot(t[:n], chirp_linear[:n], label='Linear')
plt.plot(t[:n], chirp_quadratic[:n], label='Quadrático')
plt.plot(t[:n], chirp_log[:n], label='Logarítmico')

plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Chirps no domínio do tempo')
plt.legend()
plt.grid()
plt.show()
```

## Ouvir os sinais

```python
Audio(chirp_linear, rate=fs)
```

```python
Audio(chirp_quadratic, rate=fs)
```

```python
Audio(chirp_log, rate=fs)
```

## Comentários

Os sinais chirp apresentam variação contínua de frequência ao longo do tempo.

- No chirp linear, a frequência cresce de forma uniforme entre 500 Hz e 10000 Hz.
- No chirp quadrático, a variação de frequência ocorre de forma não linear.
- No chirp logarítmico, a variação ocorre proporcionalmente em escala logarítmica.

Ao reproduzir os sinais, percebe-se que o som começa grave e termina agudo, caracterizando a varredura de frequência.
