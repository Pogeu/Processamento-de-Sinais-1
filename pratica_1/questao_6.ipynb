## Questão 6 — Convolução com resposta ao impulso

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
```

```python
fs_h, h = wavfile.read('h_banheiro.wav')
fs_t, taca = wavfile.read('sinal_taca.wav')
fs_a, audio = wavfile.read('handel.wav')
```

⚠️ Os sinais precisam ter a **mesma frequência de amostragem**.

```python
y_audio = np.convolve(audio, h)
y_taca = np.convolve(taca, h)
```

## Resultado no tempo

```python
plt.figure(figsize=(12,4))
plt.plot(y_audio)
plt.title('Áudio propagado no ambiente')
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(y_taca)
plt.title('Taça propagada no ambiente')
plt.show()
```

## Ouvir resultados

```python
Audio(y_audio, rate=fs_a)
```

```python
Audio(y_taca, rate=fs_t)
```

## Comentários

A convolução entre os sinais de entrada e a resposta ao impulso do banheiro permite simular a propagação desses sons dentro do ambiente modelado.

No domínio do tempo observa-se que os sinais resultantes apresentam maior duração em comparação aos sinais originais. Isso ocorre porque o ambiente introduz múltiplas reflexões sonoras, que se somam ao sinal direto.

Ao ouvir os sinais resultantes, percebe-se claramente a presença de **reverberação**, característica de ambientes fechados como banheiros, onde superfícies rígidas refletem fortemente as ondas sonoras.

Assim, a convolução entre um sinal e a resposta ao impulso de um ambiente permite simular realisticamente como um som seria percebido nesse espaço.
