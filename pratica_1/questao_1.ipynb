## Questão 1 — Geração e análise de sinais

Sinal analisado:
$$x(t) = cos(2\pi f t)$$

Parâmetros:
- Duração: 5 s
- Frequência de amostragem: 44.1 kHz
- Frequências testadas: 500 Hz, 5000 Hz e 10000 Hz

## Importar bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

## Definir parâmetros do sinal

```python
fs = 44100
duracao = 5

t = np.linspace(0, duracao, int(fs*duracao), endpoint=False)

frequencias = [500, 5000, 10000]
```

## Gerar os sinais

```python
sinais = {}

for f in frequencias:
    sinais[f] = np.cos(2*np.pi*f*t)
```

## Plotar sinais no domínio do tempo (primeiros 5 ms)

```python
plt.figure(figsize=(12,6))

for f in frequencias:
    x = sinais[f]
    n = int(0.005*fs)
    plt.plot(t[:n], x[:n], label=f"{f} Hz")

plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinais no domínio do tempo')
plt.legend()
plt.grid()
plt.show()
```

## Reprodução dos sinais

```python
Audio(sinais[500], rate=fs)
```

```python
Audio(sinais[5000], rate=fs)
```

```python
Audio(sinais[10000], rate=fs)
```

## Comentários

Observa-se que o aumento da frequência aumenta a quantidade de oscilações do sinal no mesmo intervalo de tempo.

- O sinal de **500 Hz** possui período maior e menos ciclos visíveis.
- O sinal de **5000 Hz** apresenta oscilações mais rápidas.
- O sinal de **10000 Hz** possui ainda mais ciclos no mesmo intervalo.

Ao reproduzir os sinais gerados, percebe-se que o aumento da frequência resulta em sons mais agudos.
