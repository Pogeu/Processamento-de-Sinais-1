# Processamento de Sinais I — Aula Prática 1
## Questão 4 — Resposta ao impulso de uma sala

Procedimento conceitual para medir a resposta ao impulso de um ambiente.

### Procedimento

1. Posicionar uma fonte sonora no ambiente (alto‑falante).
2. Posicionar um microfone no ponto de medição.
3. Emitir um sinal conhecido (normalmente um chirp ou impulso).
4. Gravar a resposta do ambiente.
5. Utilizar técnicas de deconvolução para obter a resposta ao impulso.


### Limitações

- ruído ambiente
- não linearidades do alto‑falante
- reflexões múltiplas
- posicionamento do microfone
