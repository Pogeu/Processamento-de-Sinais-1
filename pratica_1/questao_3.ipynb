## Questão 3 — Análise do arquivo handel.wav

Este notebook:
- lê o arquivo `handel.wav`
- plota o sinal no domínio do tempo
- reproduz o áudio em diferentes frequências de amostragem
- calcula o espectro do sinal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
```

```python
fs, audio = wavfile.read('handel.wav')
t = np.arange(len(audio)) / fs
```

## Gráfico no domínio do tempo

```python
plt.figure(figsize=(12,4))
plt.plot(t, audio)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal de áudio — domínio do tempo')
plt.grid()
plt.show()
```

## Reproduzir áudio

```python
Audio(audio, rate=fs)
```

### Dobro da frequência de amostragem

```python
Audio(audio, rate=2*fs)
```

### Quádruplo da frequência

```python
Audio(audio, rate=4*fs)
```

## Espectro (FFT)

```python
fft_vals = np.fft.fft(audio)
freq = np.fft.fftfreq(len(audio), 1/fs)

plt.figure(figsize=(12,4))
plt.plot(freq[:len(freq)//2], np.abs(fft_vals[:len(freq)//2]))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro do áudio')
plt.grid()
plt.show()
```
