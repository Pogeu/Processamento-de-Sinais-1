## Questão 5 — Análise de sinais acústicos

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
```

```python
fs_h, h = wavfile.read('../data/h_banheiro.wav')
fs_t, taca = wavfile.read('../data/sinal_taca.wav')
```

## Gráficos no tempo

```python
plt.figure(figsize=(12,4))
plt.plot(h)
plt.title('Resposta ao impulso — banheiro')
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(taca)
plt.title('Sinal da taça')
plt.show()
```

## Ouvir sinais

```python
Audio(h, rate=fs_h)
```

```python
Audio(taca, rate=fs_t)
```

## Comentários

O sinal **h_banheiro.wav** corresponde à resposta ao impulso de um ambiente. No domínio do tempo observa-se um pico inicial seguido por diversas reflexões de menor amplitude ao longo do tempo. Essas reflexões ocorrem devido à propagação da onda sonora e às múltiplas reflexões nas superfícies do ambiente, como paredes, piso e teto.

Já o sinal **sinal_taca.wav** apresenta uma oscilação característica de um objeto ressonante após um impacto. Observa-se um pico inicial seguido por oscilações cuja amplitude decai gradualmente ao longo do tempo, indicando dissipação de energia do sistema.

Ao ouvir os sinais, percebe-se que o som da taça apresenta uma ressonância clara e definida, enquanto o sinal **h_banheiro.wav** soa como um estalo seguido por reverberações, representando a resposta do ambiente a um impulso.
